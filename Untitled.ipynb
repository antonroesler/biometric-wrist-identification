```python
from python_speech_features import mfcc
import scipy.io.wavfile as wav
def get_file_features(wav_fname, num_ceps):
    """
    Extract mfcc features from a file.
    """
    # read wave
    fs, sig = wav.read(wav_fname)

    # get mfccs
    mfccs = mfcc(sig, samplerate=fs, winlen=0.025, winstep=0.01,
                     numcep=num_ceps, nfilt=26, nfft=512, lowfreq=0,
                     highfreq=None, preemph=0.97, ceplifter=22,
                     appendEnergy=False)

    # compute mfcc means
    mfcc_means = np.round(mfccs.mean(axis=0), 3)
    return mfcc_means 
```

```python
a = get_file_features('short2.wav', 25)
```

```python
from python_speech_features import mfcc
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["figure.figsize"] = [20, 10]
plt.rcParams["figure.autolayout"] = True

(rate, sig) = wav.read("short2.wav")
mfcc_data = mfcc(sig, rate)
fig, ax = plt.subplots()
mfcc_data = np.swapaxes(mfcc_data, 0, 1)

cax = ax.imshow(mfcc_data, interpolation='nearest', cmap='jet', origin='lower')

plt.show()
```
